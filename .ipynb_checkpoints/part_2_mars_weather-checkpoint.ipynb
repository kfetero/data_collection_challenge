```python
# Import relevant libraries
from splinter import Browser
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
import pandas as pd
```

```python
browser = Browser('chrome')
```

### Step 1: Visit the Website

Use automated browsing to visit the [Mars Temperature Data Site](https://static.bc-edx.com/data/web/mars_facts/temperature.html). Inspect the page to identify which elements to scrape.

   > **Hint** To identify which elements to scrape, you might want to inspect the page by using Chrome DevTools to discover whether the table contains usable classes.


```python
# Visit the website
# https://static.bc-edx.com/data/web/mars_facts/temperature.html
url = "https://static.bc-edx.com/data/web/mars_facts/temperature.html"
browser.visit(url)
html = browser.html
```

### Step 2: Scrape the Table

Create a Beautiful Soup object and use it to scrape the data in the HTML table.

Note that this can also be achieved by using the Pandas `read_html` function. However, use Beautiful Soup here to continue sharpening your web scraping skills.

```python
# Create a Beautiful Soup Object
soup = BeautifulSoup(html, 'html.parser')
```

```python
# Extract table data from the table in the website 
table_mars_temperature = soup.find('table', class_="table")
table_mars_temperature
```

### Step 3: Store the Data

Assemble the scraped data into a Pandas DataFrame. The columns should have the same headings as the table on the website. Heres an explanation of the column headings:

* `id`: the identification number of a single transmission from the Curiosity rover
* `terrestrial_date`: the date on Earth
* `sol`: the number of elapsed sols (Martian days) since Curiosity landed on Mars
* `ls`: the solar longitude
* `month`: the Martian month
* `min_temp`: the minimum temperature, in Celsius, of a single Martian day (sol)
* `pressure`: The atmospheric pressure at Curiosity's location

```python
# Create an empty list
columns_heading_list = []
rows_data_list = []
# Loop through the scraped data to get the headings of the table and save it in a list
for column_heading in table_mars_temperature.find_all("th"):
    # Extract all columns name from the table
    column_name = column_heading.text
    # Add colum names to the list
    columns_heading_list.append(column_name)
    
columns_heading_list
```

```python
# Create a Pandas DataFrame by using the list of rows and a list of the column names
# create dataframe with the column headings
mars_data_df = pd.DataFrame(columns = columns_heading_list)

# Loop through the scraped data to get the rows of the table and save it in a list
# Create a for loop to fill mars_data_df 
for tr_tag in table_mars_temperature.find_all('tr'):
    # get td tags
    td_data = tr_tag.find_all('td')
    #create a list
    text_data=[]
    data_count = 0
    #for loop to get the data from the <td> tags
    for row_data in td_data:
        # add data at the end of the list
        text_data.append(row_data.text)
        data_count = data_count + 1
    # print(text_data)
    length = len(mars_data_df)
    # fill data in the dataframe mars_data_df
    if data_count >= 1:
        mars_data_df.loc[length] = text_data
```

```python
# Confirm DataFrame was created successfully
mars_data_df.head()
```

### Step 4: Prepare Data for Analysis

Examine the data types that are currently associated with each column. If necessary, cast (or convert) the data to the appropriate `datetime`, `int`, or `float` data types.

  > **Hint** You can use the Pandas `astype` and `to_datetime` methods to accomplish this task.


```python
# Examine data type of each column
mars_data_df.dtypes
```

```python
# Change data types for data analysis
mars_data_df['terrestrial_date'] = pd.to_datetime(mars_data_df['terrestrial_date'], errors='coerce')
mars_data_df = mars_data_df.astype({"sol": "int64","ls": "int64","month": "int64","min_temp": "float64", "pressure": "float64" }, errors='raise')
```

```python
# Confirm type changes were successful by examining data types again
mars_data_df.dtypes
```

### Step 5: Analyse the Data

Analyse your dataset by using Pandas functions to answer the following questions:

1. How many months exist on Mars?
2. How many Martian (and not Earth) days worth of data exist in the scraped dataset?
3. What are the coldest and the warmest months on Mars (at the location of Curiosity)? To answer this question:
    * Find the average the minimum daily temperature for all of the months.
    * Plot the results as a bar chart.
4. Which months have the lowest and the highest atmospheric pressure on Mars? To answer this question:
    * Find the average the daily atmospheric pressure of all the months.
    * Plot the results as a bar chart.
5. About how many terrestrial (Earth) days exist in a Martian year? To answer this question:
    * Consider how many days elapse on Earth in the time that Mars circles the Sun once.
    * Visually estimate the result by plotting the daily minimum temperature.


```python
# 1. How many months are there on Mars?
month_grouped = mars_data_df.groupby(['month'])
month = month_grouped['month'].count()
month
```

```python
# 2. How many Martian days' worth of data are there?
martian_days = mars_data_df['sol'].count()
martian_days
```

```python
# 3. What is the average low temperature by month?
month_grouped = mars_data_df.groupby(['month'])
mean_min_temp = month_grouped['min_temp'].mean()
mean_min_temp
```

```python
# Plot the average temperature by month
# Create a bar chart from mean_min_temp (average temperature)
month_temp_chart =  mean_min_temp.plot(kind='bar', figsize=(6,4.5))

# Set the xlabel and ylabel
month_temp_chart.set_xlabel("Month")
month_temp_chart.set_ylabel("Temperature in Celsius")
plt.tight_layout()
plt.show()
```

```python
# create a new DataFrame
mean_min_temp_df = mean_min_temp.to_frame()
# sort values ascending order
mean_min_temp_df = mean_min_temp_df.sort_values(by=['min_temp'], ascending=True)
mean_min_temp_df
```

```python
# Identify the coldest and hottest months in Curiosity's location
coldest_hottest_months_chart =  mean_min_temp_df.plot(kind='bar', figsize=(6,4.5), legend=False)
# Set the xlabel and ylabel
coldest_hottest_months_chart.set_xlabel("Month")
coldest_hottest_months_chart.set_ylabel("Temperature in Celsius")
plt.tight_layout()
plt.show()
```

```python
# 4. Average pressure by Martian month
month_grouped = mars_data_df.groupby(['month'])
mean_pressure= month_grouped['pressure'].mean()
mean_pressure
```

```python
# create a new DataFrame
mean_pressure_df = mean_pressure.to_frame()
# sort values ascending order
mean_pressure_df = mean_pressure_df.sort_values(by=['pressure'], ascending=True)
mean_pressure_df
```

```python
# Plot the average pressure by month
month_pressure_chart =  mean_pressure_df.plot(kind='bar', figsize=(6,4.5), legend=False)
# Set the xlabel and ylabel
month_pressure_chart.set_xlabel("Month")
month_pressure_chart.set_ylabel("Atmospheric Pressure")
plt.tight_layout()
plt.show()
```

```python
# 5. How many terrestrial (earth) days are there in a Martian year?
#create a copy of the DataFrame
copy_mars_data_df =  mars_data_df.copy()
#add new column to the  new DataFrame
copy_mars_data_df["Number_terrestrial_days"] = pd.Series(list(range(len(copy_mars_data_df['terrestrial_date']))))
# copy_mars_data_df

terrestrial_days_in_martian_year = copy_mars_data_df.plot(x='Number_terrestrial_days', y='min_temp', legend=False)
terrestrial_days_in_martian_year.set_xlabel("Number of terrestrial days")
terrestrial_days_in_martian_year.set_ylabel("Minimum temperature")
plt.tight_layout()
plt.show()
```

On average, the third month has the coldest minimum temperature on Mars, and the eighth month is the warmest. But it is always very cold there in human terms!



Atmospheric pressure is, on average, lowest in the sixth month and highest in the ninth.

The distance from peak to peak is roughly 1425-750, or 675 days. A year on Mars appears to be about 675 days from the plot. Internet search confirms that a Mars year is equivalent to 687 earth days.
